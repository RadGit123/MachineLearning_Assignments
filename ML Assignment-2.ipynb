# outliners

```python
Q1_Age=df['age'].quantile(0.25)
Q3_Age=df['age'].quantile(0.75)

IQR=Q3_Age - Q1_Age

Lower_Bound=Q1_Age-1.3*IQR
Upper_Bound=Q3_Age+1.3*IQR

outliners=(df['age'] < Lower_Bound) | (df['age'] > Upper_Bound)
print("outliners",df[outliners])
afterOutlinerremoved_df=df[(df['age']>=Lower_Bound) & (df['age']<=Upper_Bound)]

print(afterOutlinerremoved_df)

print(f'Number of outliers in age column:{outliners.sum()}')
```

# outliners

```python
Q1_salary=df['salary'].quantile(0.25)
Q3_salary=df['salary'].quantile(0.75)

IQR=Q3_salary - Q1_salary

Lower_Bound=Q1_salary-1.3*IQR
Upper_Bound=Q3_salary+1.3*IQR

outliners=(df['salary'] < Lower_Bound) | (df['salary'] > Upper_Bound)
print("outliners",df[outliners])

afterOutlinerremoved_df=df[(df['salary']>=Lower_Bound) & (df['salary']<=Upper_Bound)]

print(afterOutlinerremoved_df)

print(f'Number of outliers in salary column:{outliners.sum()}')
```
